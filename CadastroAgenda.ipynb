```python
from google.colab import drive
drive.mount('/gdrive')

def menu():
    VoltarMenuPrincipal = 's'
    while  VoltarMenuPrincipal == 's':
      opcao= input('''
  ==================================================
              PROJETO AGENDA EM PYTHON
  MENU:

  [1] CADASTRAR CONTATO
  [2] LISTAR CONTATO
  [3] DELETAR CONTATO
  [4] BUSCAR CONTATO PELO NOME
  [5] ATUALIZAR CONTATO
  [6] SAIR
  ==================================================
  ESCOLHA UMA OPÇÃO ACIMA:
  ''')
      if opcao== "1":
        cadastrar_contato()
      elif opcao == "2":
          listar_contato()
      elif opcao == "3":
          deletar_contato()
      elif opcao == "4":
          buscar_contato()
      elif opcao == "5":
          atualizar_contato()
      else:
          sair()
          break

      VoltarMenuPrincipal=input("Deseja voltar ao menu principal? (S/N): ").lower()

def cadastrar_contato():
    idContado = input("Entre com o ID do Contato: ")
    NomeContato = input("Entre com o nome do contato: ")
    TelefoneContato = input("Entre com o telefone do contato: ")
    EmailContato = input("Entre comm o e-mail do contato: ")
    try:
        with open("/gdrive/My Drive/base_dados/agenda.txt","a") as agenda:
          dados = f'{idContado};{NomeContato};{TelefoneContato};{EmailContato} \n'
          agenda.write(dados)
        agenda.close()
        print(f'Dados gravados com sucesso!')
    except:
        print(f'ERRO na gravação de dados!')

def listar_contato():
    with open('/gdrive/My Drive/base_dados/agenda.txt','r') as agenda:
      !cat '/gdrive/My Drive/base_dados/agenda.txt'
    agenda.close()

def deletar_contato():
    nomeDeletado = input("Entre com o nome para ser excluído: ").lower()
    with open('/gdrive/My Drive/base_dados/agenda.txt', 'r', encoding='ISO-8859-1') as agenda:
      linhas = agenda.readlines()
    # Filtra as linhas que não contêm o nome a ser deletado
      linhas_filtradas = [linha for linha in linhas if nomeDeletado not in linha.lower()]

      with open('/gdrive/My Drive/base_dados/agenda.txt', 'w', encoding='ISO-8859-1') as agenda:
        agenda.writelines(linhas_filtradas)

    print(f'Contato deletado com sucesso!')
    listar_contato()

def buscar_contato():
    nome = input("Digite o nome a ser procurado: ").upper()
    with open('/gdrive/My Drive/base_dados/agenda.txt', 'r', encoding='ISO-8859-1') as agenda:
      for contato in agenda:
         if nome in contato.split(";")[1].upper():
            print(contato)
    agenda.close()

def atualizar_contato():
    nomeAlterado = input("Entre com o nome para ser alterado: ").lower()
    with open('/gdrive/My Drive/base_dados/agenda.txt', 'r', encoding='ISO-8859-1') as agenda:
      linhaAlterada = agenda.readlines()
      # Filtra as linhas que não contêm o nome a ser auterado
      linha_Alterada = [linha for linha in linhaAlterada if nomeAlterado not in linha.lower()]
      with open('/gdrive/My Drive/base_dados/agenda.txt', 'w', encoding='ISO-8859-1') as agenda:
        agenda.writelines(linha_Alterada)

    idContado = input("Entre com o ID do contato novo: ")
    NomeContato = input("Entre com o nome do contato novo: ")
    TelefoneContato = input("Entre com o telefone do contato novo: ")
    EmailContato = input("Entre comm o e-mail do contato novo: ")

    try:
      with open('/gdrive/My Drive/base_dados/agenda.txt', 'a', encoding='ISO-8859-1') as agenda:
          dados = f'{idContado};{NomeContato};{TelefoneContato};{EmailContato} \n'
          agenda.write(dados)
          agenda.close()
          print(f'Dados alterados com sucesso!')
    except:
          print(f'ERRO na alteração de dados!')

def sair():
  print(f'Até mais!!!')
  exit()

def main():
    menu()
main()

```
